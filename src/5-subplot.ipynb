### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.


import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line Chart
t = np.arange(0, 10, 1)
y = t ** 2
axs[0, 0].plot(t, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True)

# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Scatter Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x, y, color='blue')
for i, j in zip(x, y):
    axs[1, 0].annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')

# Pie Chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple'])
axs[1, 1].set_title('Programming Language Popularity')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
