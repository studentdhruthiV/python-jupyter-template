### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

# Define data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plt.scatter(x, y, color='blue')

# Annotate each point
for i, j in zip(x, y):
    plt.annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(5,5), ha='center')

# Add title
plt.title('Scatter Plot with Annotations')

# Label axes
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
```
